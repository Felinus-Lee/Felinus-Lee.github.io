```python
# 定义图的邻接表
graph = {
    'A': [('B', 4), ('C', 2), ('D', 5)],
    'B': [('A', 4), ('C', 1), ('E', 1)],
    'C': [('A', 2), ('B', 1), ('D', 3)],
    'D': [('A', 5), ('C', 3), ('E', 2)],
    'E': [('B', 1), ('D', 2)]
}

# 打印邻接表
def print_adj_list(graph):
    for node, edges in graph.items():
        edge_list = ', '.join(f'{neighbor} (weight {weight})' for neighbor, weight in edges)
        print(f'{node}: {edge_list}')

# 打印图的结构
print("Graph adjacency list:")
print_adj_list(graph)
```
