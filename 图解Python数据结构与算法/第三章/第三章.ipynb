```python
def factorial(n):
    # 基本情况：当 n 为 0 时，阶乘为 1
    if n == 0:
        print(f"Base case: factorial({n}) = 1")
        return 1
    # 递归情况：n! = n * (n-1)!
    else:
        result = n * factorial(n - 1)
        print(f"factorial({n}) = {n} * factorial({n-1}) = {result}")
        return result

# 计算 5 的阶乘
result = factorial(5)
print("\n5 的阶乘是:", result)
```

```python
def greedy_coin_change(coins, amount):
    coins.sort(reverse=True)  # 按面额从大到小排序
    num_coins = {coin: 0 for coin in coins}

    for coin in coins:
        while amount >= coin:
            amount -= coin
            num_coins[coin] += 1

    return num_coins if amount == 0 else None

coins = [1, 2, 5]
amount = 13
result = greedy_coin_change(coins, amount)

if result:
    print("需要的硬币数量：", result)
else:
    print("无法凑成目标金额")
```
